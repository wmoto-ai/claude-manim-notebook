````python
import os
import anthropic
import random
import time
from IPython.core.magic import register_cell_magic

client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
prompt_w_history = []

@register_cell_magic
def _(line, cell):
    system_prompt = f"""あなたはJupyterNotebookを使うPythonデータサイエンティストであり、Manimアニメーションの専門家です。
    入力された指示をもとに、Manimを使用したPythonアニメーションプログラムを作成してください。
    コードはIPython magic command %%manim を使用して、Jupyter Notebook上で直接実行できるようにしてください。
    アニメーションを作成する際は、以下の点に注意してください:
    1. アニメーションに必要なオブジェクト（テキスト、図形など）を適切に作成し、配置してください。
    2. アニメーションの流れを適切に設計し、シーンの遷移がスムーズになるようにしてください。
    3. Manimの各種メソッド（Write, FadeIn, FadeOut, Rotateなど）を効果的に使用し、視覚的に魅力的なアニメーションを実現してください。
    4. マジックコマンドの引数は -qm（中品質、エラーのみ表示）とクラス名のみを使用し、他は使用しないでください
    5. コード内にはアニメーションの概要や各ステップの説明をコメントとして適切に記載してください。
    回答には、1つのPythonコードブロック（マジックコマンド、クラス定義、コンストラクタを含む）のみを含め```pythonで囲ってください。
    """

    user_prompt = cell
    filename = line

    if filename != "":
        system_prompt += f"""
        # 入力データ:ファイルパス
        {filename}
        なお、データはUTF-8で読み込んでください。
        """

    messages = [
        {"role": "user", "content": user_prompt}
    ]

    prompt_w_history.extend(messages)

    try:
        response = client.messages.create(
            max_tokens=2048,
            messages=prompt_w_history,
            model="claude-3-opus-20240229",
            system=system_prompt
        )
    except Exception as e:
        print(e)
        time.sleep(3)
        response = client.messages.create(
            max_tokens=2048,
            messages=prompt_w_history,
            model="claude-3-opus-20240229",
            system=system_prompt
        )

    assistant_response = response.content[0].text

    print("APIからの応答:", assistant_response)

    python = ""
    if "```python" in assistant_response and "```" in assistant_response:
        python = assistant_response.split("```python")[1].split("```")[0]

    print("抽出されたPythonコード:", python)

    out = get_ipython().run_cell(python)
    prompt_w_history.append({"role": "assistant", "content": str(assistant_response) + '\n' + str(out.result)})
````
