````python
import os
import openai
import time
from dotenv import load_dotenv, find_dotenv
from IPython.core.magic import register_cell_magic
_ = load_dotenv(find_dotenv())

openai.api_key  = os.getenv('OPENAI_API_KEY')

def get_completion_from_messages(messages,temperature=0):
    response = openai.ChatCompletion.create(
        model="gpt-4-1106-preview",
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message["content"]

prompt_w_history = []

@register_cell_magic
def _(line, cell):
    system_prompt = f"""あなたはJupyterNotebookを使うPythonデータサイエンティストであり、Manimアニメーションの専門家です。
    入力された指示をもとに、Manimを使用したPythonアニメーションプログラムを作成してください。
    コードは!!必ずIPython magic command %%manim から始めて!!、Jupyter Notebook上で直接実行できるようにしてください。
    アニメーションを作成する際は、以下の点に注意してください:
    1. アニメーションに必要なオブジェクト（テキスト、図形など）を適切に作成し、配置してください。
    2. アニメーションの流れを適切に設計し、シーンの遷移がスムーズになるようにしてください。
    3. Manimの各種メソッド（Write, FadeIn, FadeOut, Rotateなど）を効果的に使用し、視覚的に魅力的なアニメーションを実現してください。
    4. マジックコマンドの引数は **-qm（中品質、エラーのみ表示）だけ**を使用してください
    5. コード内にはアニメーションの概要や各ステップの説明をコメントとして適切に記載してください。
    6. 別途画像ファイルを扱う際にはユーザーからのファイル指示があった場合のみとすること
    回答には、1つのPythonコードブロック（マジックコマンド、クラス定義、コンストラクタを含む）のみを含めてください。
    """

    user_prompt = cell
    filename = line

    if filename != "":
        system_prompt += f"""
# 入力データ:ファイルパス
{filename}
なお、データはUTF-8で読み込んでください。
"""

    messages =  [
    {'role':'system', 'content':system_prompt},
    ]

    prompt_w_history.append({'role':'user', 'content':user_prompt})

    messages.extend(prompt_w_history)

    response = get_completion_from_messages(messages, temperature=0)

    print(response)

    python = response.split("```python")[1].split("```")[0]

    out = get_ipython().run_cell(python)
    prompt_w_history.append({'role':'assistant', 'content':response+'\n'+str(out.result)})

````
